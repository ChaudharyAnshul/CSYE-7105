### Import statements

```python
import pandas as pd
from nycflights13 import flights
from dask.distributed import Client
```

#### 1. Remove the samples with “NaN” in the feature “dep_delay” in this dataframe.    (8 pts)

```python
# Convert flights data into a Pandas DataFrame
flights_df = flights
```

```python
# Check Type
type(flights_df)
```

```python
# display Head
flights_df.head()
```

```python
# Remove rows with NaN in the 'dep_delay' column
flights_df_cleaned = flights_df.dropna(subset=['dep_delay'])
```

#### 2. Start a Dask Client and set “n_workers=4” for this client.    (4 pts)

```python
client = Client(n_workers=4)
client
```

```python
import dask.dataframe as dd

# Convert the cleaned Pandas DataFrame to a Dask DataFrame
flights_dask_df = dd.from_pandas(flights_df_cleaned, npartitions=4)

# Calculate mean and standard deviation of 'dep_delay'
dep_delay_mean = flights_dask_df['dep_delay'].mean().compute()
dep_delay_std = flights_dask_df['dep_delay'].std().compute()

# Print the results
print("Mean of dep_delay:", dep_delay_mean)
print("Standard deviation of dep_delay:", dep_delay_std)
```

```python
from dask.distributed import performance_report

with performance_report(filename="dask_report_3.html"):
  # Calculate mean and standard deviation of 'dep_delay'
  dep_delay_mean = flights_dask_df['dep_delay'].mean().compute()
  dep_delay_std = flights_dask_df['dep_delay'].std().compute()
  
  # Print the results
  print("Mean of dep_delay:", dep_delay_mean)
  print("Standard deviation of dep_delay:", dep_delay_std)
```

```python
client.close()
```
